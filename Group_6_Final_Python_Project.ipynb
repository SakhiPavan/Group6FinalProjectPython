For this project we will be exploring Titanic Dataset. This dataset provides information about passengers who were aboard the RMS Titanic during its ill-fated maiden voyage. This dataset is often used for exploring patterns and factors associated with survival on the Titanic.


The dataset includes the following columns:

PassengerId: Unique identifier for each passenger.

Survived: Survival status of the passenger (0 = Not Survived, 1 = Survived).

Pclass: Passenger class (1 = First class, 2 = Second class, 3 = Third class).

Sex: Gender of the passenger.

Age: Age of the passenger.

SibSp: Number of siblings/spouses aboard the Titanic.

Parch: Number of parents/children aboard the Titanic.

Fare: Fare paid by the passenger.

Embarked: Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton).

# Setting up the Data

**Let's get ready to set up our data for our testing models.**

#1. Decision Tree:

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin', axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

## Train Test Split

**Now its time to split our data into a training set and a testing set!**

**Using sklearn to split our data into a training set and a testing set.**

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
```

```python
# Decide variables to use
X = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
y = df['Survived']
```

```python
# train/test split (80/20)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=65)
```

```python
# Train a decision tree
dt = DecisionTreeClassifier(max_depth=3)
dt_model = dt.fit(x_train, y_train)
```

## Predictions and Evaluation of Decision Tree
**Creating predictions from the test set and creating a classification report and a confusion matrix.**

```python
# Predict the value

dt_preds = dt.predict(x_test)
```

```python
# Check accuracy score

from sklearn.metrics import accuracy_score
dt_accuracy = accuracy_score(y_test, dt_preds)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, dt_preds))
```

```python
print(confusion_matrix(y_test, dt_preds))
```

```python
# Tree visualization
fig = plt.figure(figsize=(12, 10))
tree.plot_tree(dt_model, feature_names=list(X.columns), class_names=['Not survived', 'Survived'])
plt.show()
```

```python
# Evaluate model performance on testing set
print("Accuracy score on Testing set: ", dt_model.score(x_test, y_test))
```

#2. Random Forest

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Train a random forest
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state=42)
rfc.fit(x_train, y_train)
```

```python
# Make Predictions

rfc_preds = rfc.predict(x_test)
```

```python
# Check the accuracy score

rfc_accuracy = accuracy_score(y_test, rfc_preds)
```

```python
print(classification_report(y_test, rfc_preds))
```

```python
# Evaluate model performance on testing set
print("Accuracy score on Testing set: ", rfc.score(x_test, y_test))
```

#3. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

logistic_reg = LogisticRegression(max_iter=1000)
```

```python
# Train the models
logistic_reg.fit(x_train, y_train)
```

```python
# Make predictions
logistic_reg_preds = logistic_reg.predict(x_test)
```

```python
# Evaluate accuracy
logistic_reg_accuracy = accuracy_score(y_test, logistic_reg_preds)


print(f"Logistic Regression Accuracy: {logistic_reg_accuracy}")
```

```python
print(classification_report(y_test, logistic_reg_preds))
```

```python
# Creating a bar plot for model comparison
models = ['Logistic Regression', 'Decision Tree', 'Random Forest']
accuracies = [logistic_reg_accuracy, dt_accuracy, rfc_accuracy]

plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color='orange')
plt.title('Accuracy of Different Models')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.ylim(0.6, 1.0)
plt.show()
```

```python
# Creating confusion matrix for each model

from sklearn.metrics import confusion_matrix

# Make predictions
models = [logistic_reg, dt, rfc]
model_names = ['Logistic Regression', 'Decision Tree', 'Random Forest']
predictions = []

for model in models:
    predictions.append(model.predict(x_test))

plt.figure(figsize=(16, 12))
for i in range(len(models)):
    plt.subplot(2, 2, i+1)
    cm = confusion_matrix(y_test, predictions[i])
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', cbar=False)
    plt.title(f'Confusion Matrix - {model_names[i]}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
plt.tight_layout()
plt.show()
```
